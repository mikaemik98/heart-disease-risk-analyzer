```python
#import libraries neede for data loading and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("Analysis module loaded successfully.")
```

```python
#define column names since the data file has no headers
columns = ['age', 'sex', 'cp', 'trestbps', 'cholesterol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

#load dataset
#na_values='?' tells pandas to treat '?' as missing data (NaN)
data = pd.read_csv('data/processed.cleveland.data', names=columns, na_values='?')

#ensure cholesterol is treated as a number, invalid values become missing
data['cholesterol'] = pd.to_numeric(data['cholesterol'], errors='coerce')

#remove rows with missing cholesterol for clean visualizations
data_clean = data.dropna(subset=['cholesterol'])

#quick overview of the dataset structure and first 5 rows
print(data_clean.head())
```

```python
#summary statistics for all columns
#shows count, mean, min, max for each measurement
print(data_clean.describe())

#check how many values are missing per column
print("\nMissing values: \n", data_clean.isnull().sum())

#count how many patients have heaert disease vs not
#target > 0 means heaert disease present
print("\nHeart Disease cases:", data_clean['target'].value_counts().to_dict())
```

```python
###data = data.dropna()

#scatter plot of age vs cholesterol colored by heart disease status
#red dots = heart diseasem blue dots = no heart disease
plt.scatter(data_clean['age'], data_clean['cholesterol'], c=data_clean['target'], cmap='coolwarm', alpha=0.6)

plt.colorbar(label='Heart Disease Presence (0=No, 1=Yes)')

plt.xlabel('Age')
plt.ylabel('Cholesterol')

plt.title('Age vs Cholesterol')

plt.show()
#look for patterns in the data
```

```python
#age distribution of all patients in the dataset
#shows which age groups are most represented
data_clean['age'].hist(bins=20, color='steelblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
#cholesterol distribution across all patients
#normal cholesterol is 200mg/dl
#high risk is above 240mg/dl
data_clean['cholesterol'].hist(bins=20, color='steelblue', edgecolor='black')
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol (mg/dl)')
plt.ylabel('Number of Patients')
plt.show()
```

```python
#age vs maximum heart rate colored by heart disease status
#healthy hearts should reach higher maximum heart rates during exercise
#lower maximum heart rate often indicates heart disease
plt.scatter(data_clean['age'], data_clean['thalach'], c=data_clean['target'], cmap='coolwarm', alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (Thalach)')
plt.title('Age vs Thalach')
plt.colorbar(label='Heart Disease Presence (0=No, 1=Yes)')
plt.show()
```

```python
#count of patients with and without heart disease
#shows if the dataset is balanced between the two groups
#imbalanced data can make models biased toward the majority group
labels = ['No Heart Disease', 'Heart Disease']
counts = (data_clean['target'] > 0).astype(int).value_counts().sort_index()
plt.bar(labels, counts, color=['steelblue', 'red'])
plt.title('Distribution of Heart Disease Cases')
plt.ylabel('Number of Patients')
plt.xlabel('Heart Disease Status')
plt.show()
```
