```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("Analysis module loaded successfully.")
```

```python
columns = ['age', 'sex', 'cp', 'trestbps', 'cholesterol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

data = pd.read_csv('data/processed.cleveland.data', names=columns, na_values='?')

data['cholesterol'] = pd.to_numeric(data['cholesterol'], errors='coerce')

data_clean = data.dropna(subset=['cholesterol'])

print(data_clean.head())
```

```python
print(data_clean.describe())
print("\nMissing values: \n", data_clean.isnull().sum())
print("\nHeart Disease cases:", data_clean['target'].value_counts().to_dict())
```

```python
###data = data.dropna()

plt.scatter(data_clean['age'], data_clean['cholesterol'], c=data_clean['target'], cmap='coolwarm', alpha=0.6)

plt.colorbar(label='Heart Disease Presence (0=No, 1=Yes)')

plt.xlabel('Age')
plt.ylabel('Cholesterol')

plt.title('Age vs Cholesterol')

plt.show()
```

```python
data_clean['age'].hist()
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
data_clean['cholesterol'].hist()
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.show()
```
